#### Perfect 5th Interval Data visualization

Perfect 5th Interval is known as the golden Interval in music.
It creates the most harmonic and pure sound. Such as the open strings in the string instruments, violin, viola and Cello. It is known as E,A,D,G or mi, la, re, sol. We will show the frequency plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot A as 440 HZ
x= np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("0 to 4pi")
plt.ylabel("A(440hz)")
plt.title("Sound Wave of A")
plt.show()
```

```python
#plot A 440HZ and E 699Hz
z=np.sin(1.5*x)
plt.plot(x,y,x,z)
plt.xlabel("0 to 4pi")
plt.ylabel("A(440hz), E(699hz)")
plt.title("Sound Wave of A and E")
plt.legend(["A string", "E string"])
plt.show()
```

```python

# The perfect fifth sound wave
y1=np.sin(x)+ np.sin(1.5*x)
plt.plot(x,y1, color="pink")
plt.title("Perfect 5th Sound Wave")
plt.show()
```
